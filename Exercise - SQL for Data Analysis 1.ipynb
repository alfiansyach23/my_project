## Case 5
Kamu diminta untuk mengumpulkan data yang berisi informasi mengenai *station_number, wind_speed, temperature, serta rain* untuk kemudian akan kamu analisa. Karena kamu takut kurang teliti, kamu berencana untuk menulis nama kolom dalam format vertikal agar detail penulisan dapat terlihat. Bagaimana cara mendapatkan informasi tersebut?

```python
# Menulis query dalam format vertikal
station_data = pd.read_sql_query('''SELECT 
                                        station_number,
                                        wind_speed,
                                        temperature,
                                        rain
                                    FROM 
                                        STATION_DATA''', 
                                conn_data)
station_data
```

## Case 7
Kamu diminta untuk memberikan 5 data teratas yang berisi informasi mengenai *report_code, month, day, thunder, dan tornado*. Data yang kamu berikan akan membantu pusat informasi dalam station untuk memberikan peringatan pada para penumpang. Bagaimana cara mendapatkan informasi tersebut?

```python
# Menampilan 5 data teratas mengenai kolom report_code, month, day, thunder, dan tornado
station_data = pd.read_sql_query('''SELECT 
                                        report_code,
                                        month,
                                        day,
                                        thunder,
                                        tornado
                                    FROM 
                                        STATION_DATA
                                    LIMIT
                                        5''', 
                                conn_data)
station_data
```

## Case 8
Kapan terdapat laporan mengenai adanya thunder atau tornado ? urutkan waktu kejadian dari yang terbaru

```python
# Menampilan laporan mengenai adanya thunder dan tornado dengan kejadian yang terbaru
station_data = pd.read_sql_query('''SELECT 
                                        *
                                    FROM 
                                        STATION_DATA
                                    WHERE 
                                        thunder OR
                                        tornado
                                    ORDER BY
                                        year DESC,
                                        month DESC,
                                        day DESC''', 
                                conn_data)
station_data
```

## Case 9
Berapa kali terjadinya tornado dan rata-rata wind_speed & temperatur setiap tahunnya? Tahun berapakah kejadian tornado terbanyak?

```python
# Menampilan banyaknya tornado tiap tahun dan rata-rata wind_speed dan temperatur tiap tahun
station_data = pd.read_sql_query('''
                                    SELECT 
                                        COUNT(tornado) AS Total_Tornado,
                                        AVG(wind_speed) AS Avg_Wind_Speed,
                                        AVG(temperature) AS Avg_Temperature,
                                        year
                                    FROM 
                                        STATION_DATA
                                    GROUP BY
                                        year
                                    ORDER BY
                                        tornado DESC''',
                                conn_data)
station_data
```
